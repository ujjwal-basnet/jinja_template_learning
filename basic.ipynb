from_string is a method of the Environment object that lets you create a template directly from a string, instead of loading it from a file.



### normally

```python
from write_message import max_score
from write_message import test_name
from jinja2 import Environment, FileSystemLoader

env = Environment(loader=FileSystemLoader('templates'))
template = env.get_template('message.txt')
output = template.render(name="Ujjwal" ,test_name= "english", max_score = 100, score= 90)
print(output)
```

### with from string we can skip the file and define the template inline

```python
from jinja2 import Environment

env = Environment()

# Create template directly from a string
template = env.from_string("Hello {{ name }}, your score in {{ test_name }} is {{ score }} out of {{ max_score }}")

# Render the template with data
output = template.render(name="Ujjwal" ,test_name= "english", max_score = 100, score= 90)

print(output)
```
