```python
### based on students marks we are writing if condition template
message= """Hello {{ name }}!

{% if score > 80 %}
I'm happy to inform you that you did very well on today's {{ test_name }}.
{% else %}
I'm sorry to inform you that you did not do so well on today's {{ test_name }}.
{% endif %}
You reached {{ score }} out of {{ max_score }} points.

See you tomorrow!
Anke

"""
```

```python
from write_message import max_score
from jinja2 import Environment

env = Environment()

# Create template directly from a string
template = env.from_string(message)
# Render the template with data
output = template.render(name="Ujjwal" ,test_name= "english", max_score = 100, score= 90 )
output2= template.render(name="rakesh", test_name="english", max_score=100, score= 20 )

print(output)
print("\n\n******************************")
print(output2)
```
